```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
def file_len(fname):
    with open(fname) as f:
        for i, l in enumerate(f):
            pass
    return i + 1
```

```python
def ans(filename):
    f = open(filename,'r')
    lst = []
    for line in f:
        lst.append(line)
    
    temp = []
    hum = []
    for i in range(len(lst)):
        if i%2 == 0:
            temp.append(lst[i])
        else:
            hum.append(lst[i])

    temp = [float(x[13:18]) for x in temp]
    hum = [float(x[10:15]) for x in hum]
    
    return [temp,hum]

def sizex(filename):
    return np.linspace(0,300,file_len(filename)/2)
```

```python
plt.plot(sizex('cloth.txt'),ans('cloth.txt')[0])
```

```python
plt.plot(sizex('without.txt'),ans('without.txt')[0])
```

```python
plt.plot(sizex('without.txt'),ans('without.txt')[0],label = "without cloth")
plt.plot(sizex('cloth.txt'),ans('cloth.txt')[0],label = "polyester")
plt.plot(sizex('cloth2.txt'),ans('cloth2.txt')[0],label="Wool 80% and Cashmere 20%")
plt.plot(sizex('cloth3.txt'),ans('cloth3.txt')[0],label = "Chiffon(fabric)")
plt.xlabel('seconds(s)')
plt.ylabel('Degree(C)')
plt.legend()
plt.grid()
```
